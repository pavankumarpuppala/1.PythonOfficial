## Shallow(assigning)
- assigning another ref variable

```python
x=[10,20,30,40]
y=x
print(id(x))
print(id(y))
```

```python
from PIL import Image
Image.open("E:/code/1.PythonOfficial/img/list3.png")
```

- so if we are trying to update any value in the list by using 'y' ref variable it will reflect to the x object 

```python
x=[10,20,30,40]
y=x
y[1]=777
print(x)
print(y)
print(id(x))
print(id(y))
```

- to overcome the above problem we go for cloning

### cloning: creating an exact copy of an object
- slice operator
- copy() function
- list()

#### slice operator

```python
from PIL import Image
Image.open("E:/code/1.PythonOfficial/img/list4.png")
```

```python
x=[10,20,30,40]
y=x[:] 
```

```python
y[1]=777
```

```python
print(x)
print(y)
print(id(x))
print(id(y))
```

#### copy()

```python
x=[10,20,30,40]
y=x.copy()
```

```python
y[2]=333
print(x)
print(y)
print(id(x))
print(id(y))
```

```python
a = [10,20,30,40]
b = list(a)
print(id(a))
print(id(b))

b[2]=300
print(a)
print(b)
```

<b>Using the method deepcopy from the module copy</b>

```python
from PIL import Image
Image.open("E:/code/1.PythonOfficial/img/deep1.png")
```

```python
from PIL import Image
Image.open("E:/code/1.PythonOfficial/img/deep2.png")
```

```python
from copy import deepcopy

lst1 = ['a','b',['ab','ba']]
lst2 = deepcopy(lst1)

print(id(lst1))
print(id(lst2))
print(lst1, lst2)
```

```python
lst2[0] = 'c'
print(lst1, lst2)
```

```python
lst2[2][1]='d'
print(lst1, lst2)
```

```python
print(id(lst1[0]))
print(id(lst2[0]))
```

```python
print(id(lst1[1]))
print(id(lst2[1]))
```

```python
print(id(lst1[2]))
print(id(lst2[2]))
```
